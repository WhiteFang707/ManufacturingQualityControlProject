```python
mask = df['units_produced'].between(0, 2000)
```

```python
total = len(mask)
bad = (~mask).sum()
good = mask.sum()

print(f'Total: {total}, out of range: {bad}, valid values: {good}')
```

```python
df[~mask]
```

Cleaning steps :
1) Deleting duplicates rows.
2) Changing out of range units.
The defect_columns has no corruption found therefore we can find the mean percentage of defect rate and use this information to infer the value of the units produced.
3) Formatting the date column.
4) Formatting the inspector column.

```python
# changing the out of values of units_produced
mean_defect_rate_ = df[mask].groupby('line').apply(lambda x: (x['defects_found'] / x['units_produced']).mean())
df.loc[~mask, 'units_produced'] = df.loc[~mask].apply(lambda row: round(row['defects_found']/ mean_defect_rate_[row['line']]), axis=1)
```

```python
df[~mask]
```

```python
#Formatting the date column
date_mask = df['date_parsed'].isna()
```

```python
df.loc[date_mask, 'date'] = pd.to_datetime(df.loc[date_mask, 'date'], format='mixed', dayfirst=True).dt.strftime('%d/%m/%Y')
df[date_mask]
```

```python
date_parsed = pd.to_datetime(df['date'], format='%d/%m/%Y', errors='coerce')
print(f"Remaining bad dates: {date_parsed.isnull().sum()}")
```

```python
df = df.drop(columns='date_parsed')
df
```

```python
#deleting duplicated rows
df = df.drop_duplicates().reset_index(drop=True)
df
```

```python
#Formating the inspector columns
df['inspector'] = df['inspector'].str.strip()
df['inspector'].nunique(), df['inspector'].unique()
```

```python
#saving the clean file
output_path = Path('') / 'data' / 'cleaned.csv'
df.to_csv(output_path, index=False)
```
